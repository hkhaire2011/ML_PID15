```python
import numpy as np 
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('./titanic.csv')
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
result = np.array(['zero'])
for i in range(1,19):
    result= np.append(result,'zero.'+str(i))
result
    
```

```python
data_mod = data.drop(result,axis=1)
```

```python
data_mod.describe()
data_mod=data_mod.dropna()
data_mod
```

```python
data_mod.isnull().sum()
```

```python
X = data_mod.iloc[:,1:-3].values 
y = data_mod.iloc[:,8]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
X_train = StandardScaler().fit_transform(X_train)
X_test = StandardScaler().fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_test,y_pred)
ac
```

```python
from sklearn.ensemble import RandomForestClassifier
rm = RandomForestClassifier()
rm.fit(X_train,y_train)
y_pred1 = rm.predict(X_test)
y_pred1
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred1)
cm
```

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_test,y_pred1)
ac
```

```python
rm.fit(X,y)
result = rm.predict([[24,512,1,1,0]])
result
if result == 0:
    print("not survived")
else:
    print("survived")
```
